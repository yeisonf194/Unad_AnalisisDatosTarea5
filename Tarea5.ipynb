```python
def autolabel(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate('{}'.format(height),
            xy=(bar.get_x() + bar.get_width() / 2,height),
            xytext=(0,3),
            textcoords="offset points",
            ha='center', va='bottom')
```

```python
def funcPie(values):
    val = iter(values)
    return lambda  pct: f"{pct:.1f}% ({next(val)})"
```

```python
fig, ax=plt.subplots(1,3,figsize = (16,7))
ax[0].pie(intervaloEdad1['Age'].to_list(), labels=intervaloEdad1.index.to_list(),
    autopct=funcPie(intervaloEdad1['Age'].to_list()), shadow=True, startangle=90)
ax[0].axis('equal')
ax[0].set_title('Edades Menores a 18')
ax[1].pie(intervaloEdad2['Age'].to_list(), labels=intervaloEdad2.index.to_list(),
    autopct=funcPie(intervaloEdad2['Age'].to_list()), shadow=True, startangle=90)
ax[1].axis('equal')
ax[1].set_title('Edades mayores o iguales a 18 y menores o iguales a 50')
ax[2].pie(intervaloEdad3['Age'].to_list(), labels=intervaloEdad3.index.to_list(),
    autopct=funcPie(intervaloEdad3['Age'].to_list()), shadow=True, startangle=90)
ax[2].axis('equal')
ax[2].set_title('Edades Mayores a 50')
plt.legend()
plt.show()
```

```python
familias=[]
for i in datos['Name']:
    apellido=str(i).split(',')[0]
    f=datos[datos['Name'].str.match(apellido+ r'\b',case=False)]['Name'].to_list()
    if(familias.count(f)==0):
        familias.append(f)
    familias[:5]
```

```python
len(familias)
```

```python
for i in range (0 ,len(familias)-1):
    for j in range(i+1, len(familias)):
        if len(familias[j])>len(familias[i]):
            aux=familias[i]
            familias[i]=familias[j]
            familias[j]=aux
        familias[:5]
```

```python
cont=[]
nombreFamilias=[]
for i in familias [:3]:
    aux=""
    cont.append(len(i))
    for j in i:
        aux+=j+'\n'
    nombreFamilias.append(aux)
    
fig, ax=plt.subplots(figsize = (18,7))
ax.set_ylabel('Total')
ax.set_title('Familias mas Numerosas')
bar1=ax.bar(nombreFamilias,cont)
autolabel(bar1)
plt.show()
```

```python
import seaborn as sns

bins = [0, 18, 60, 100]  # Rangos de edad
labels = ['Menor de 18', 'Entre 19 y 60', 'Mayor que 60']  # Etiquetas para cada segmento de edad
datos['AgeSegment'] = pd.cut(datos['Age'], bins=bins, labels=labels)

# Graficar la segmentación por edad y clase
sns.countplot(x='AgeSegment', hue='Pclass', data=datos)
plt.xlabel('Segmento de Edad')
plt.ylabel('Cantidad')
plt.title('Segmentación de las clases por segmento de edad')
plt.legend(title='Clase')
plt.show()
```

```python
import seaborn as sns

# Crear una nueva columna para segmentar la edad
bins = [0, 18, 59, 100]  # Rangos de edad
labels = ['Menor que 19', 'Entre 20 y 59', 'Mayor que 59']  # Etiquetas para cada segmento de edad
datos['AgeSegment'] = pd.cut(datos['Age'], bins=bins, labels=labels)

# Agrupar los datos por segmento de edad y clase, y contar el número de sobrevivientes y fallecidos
survival_counts = datos.groupby(['AgeSegment', 'Pclass', 'Survived']).size().unstack()

# Graficar el número de sobrevivientes y fallecidos por segmento de edad y clase
colors = ['red', 'blue']
survival_counts.plot(kind='bar', stacked=True, color=colors)
plt.xlabel('Segmento de Edad')
plt.ylabel('Cantidad')
plt.title('Relación entre rango de edades, clase y número de sobrevivientes/fallecidos')
plt.legend(['Fallecidos', 'Sobrevivientes'])
plt.show()
```

```python
import seaborn as sns

# Crear una tabla de contingencia de fallecidos por clase
tabla_contingencia = pd.crosstab(datos['Pclass'], datos['Survived'] == 0)

# Definir el gradiente de colores personalizado
cmap = sns.color_palette("YlOrRd", as_cmap=True)

# Graficar el heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(tabla_contingencia, cmap=cmap, annot=True, fmt='d', cbar=True)
plt.xlabel('Fallecidos')
plt.ylabel('Clase')
plt.title('Gráfico de calor de número de fallecidos por clase')
plt.show()
```
